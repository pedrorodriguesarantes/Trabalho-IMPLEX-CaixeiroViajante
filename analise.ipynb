```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('Resultados/HillClimbing - att48.tsp.txt - 10.csv', sep = ";")

fig, ax = plt.subplots()
ax.plot(df['Unnamed: 0'], df['MELHOR GLOBAL'], linewidth=2.0, label = "Melhor Global")
ax.plot(df['Unnamed: 0'], df['MELHOR LOCAL'], linewidth=2.0, label = "Melhor Local")
plt.legend()

plt.title("Hill Climbing - Att48")
plt.show()

print(min(df['MELHOR GLOBAL'][1:]))
```
